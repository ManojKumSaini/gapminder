```python
def to_tidy(df, value_name):
    tidy = (
        df.melt(id_vars=["country"],var_name="year", value_name=value_name)
        
    )
    tidy["year"] = tidy["year"].astype(int)
    return tidy
```

```python
tidy_pop  = to_tidy(pop,  "population")
tidy_life = to_tidy(life, "lifeExp")
tidy_gni  = to_tidy(gni,  "gniPPP")
```

```python
def missing_val(df, name):
    print(f"Dataset: {name}")
    miss = df.isnull().sum()
    print(f"Missing value: {miss}")
    print("-"*15)
```

```python
missing_val(tidy_pop, "Population")
missing_val(tidy_life, "Life exp")
missing_val(tidy_gni, "GNI:")
```

```python
def infor(df):
    print(df.dtypes)
    print("-"*15)    
```

```python
infor(tidy_gni)
infor(tidy_life)
infor(tidy_pop)
```

```python
tidy_pop = tidy_pop.sort_values(['country', 'year'])
tidy_pop['population'] = tidy_pop.groupby('country')['population'].ffill()
tidy_pop['population'] = tidy_pop.groupby('country')['population'].bfill()

tidy_life = tidy_life.sort_values(['country', 'year'])
tidy_life['lifeExp'] = tidy_life.groupby('country')['lifeExp'].ffill()
tidy_life['lifeExp'] = tidy_life.groupby('country')['lifeExp'].bfill()

tidy_gni = tidy_gni.sort_values(['country', 'year'])
tidy_gni['gniPPP'] = tidy_gni.groupby('country')['gniPPP'].ffill()
tidy_gni['gniPPP'] = tidy_gni.groupby('country')['gniPPP'].bfill()
```

```python
print("Checking after forward fill")
missing_val(tidy_pop, "Population")
missing_val(tidy_life, "Life exp")
missing_val(tidy_gni, "GNI:")
```

```python
df_merged = (
    tidy_pop
      .merge(tidy_life, on=["country","year"])
      .merge(tidy_gni,  on=["country","year"]))
```

```python
def convert_abbreviated_numbers(val):
    if isinstance(val, str):
        val = val.strip()
        if val.endswith("k"):
            return float(val[:-1]) * 1_000
        elif val.endswith("M"):
            return float(val[:-1]) * 1_000_000
        elif val.endswith("B"):
            return float(val[:-1]) * 1_000_000_000
    return val  

df_merged['population'] = df_merged['population'].apply(convert_abbreviated_numbers).astype(int)
df_merged['gniPPP'] = df_merged['gniPPP'].apply(convert_abbreviated_numbers).astype(int)
df_merged['lifeExp'] = df_merged['lifeExp'].astype(int)
df_merged['gniPPP'] = df_merged['gniPPP'].astype(int)

```
